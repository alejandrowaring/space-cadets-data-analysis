```python

success_fail_df = global_space_launches[['Company Name', 'Status Mission',]]
```

```python
successful_df = success_fail_df[success_fail_df['Status Mission'] == 'Success']
successful_df = successful_df.groupby('Company Name').count()

success = successful_df.rename(columns={"Status Mission" : "Status Success"})
success.head()
```

```python
failure_df = success_fail_df[success_fail_df['Status Mission'] == 'Failure']
failure_df = failure_df.groupby('Company Name').count()

fail = failure_df.rename(columns={"Status Mission" : "Status Failure"})
fail.head()
```

```python
pecentage_success_fail = pd.merge(success, fail, on="Company Name", how='outer')

pecentage_success_fail = pecentage_success_fail.fillna(0)

pecentage_success_fail["Status Success"] = pecentage_success_fail ['Status Success'].astype('int')
pecentage_success_fail['Status Failure'] = pecentage_success_fail ['Status Failure'].astype('int')

pecentage_success_fail['Missions'] = pecentage_success_fail['Status Success']+ pecentage_success_fail ['Status Failure']

pecentage_success_fail['Pecentage Successful'] = pecentage_success_fail ['Status Success'] / pecentage_success_fail ['Missions']
pecentage_success_fail['Pecentage Failure'] = pecentage_success_fail ['Status Failure'] / pecentage_success_fail ['Missions']


pecentage_success_fail.sort_values(by=['Status Success'], inplace=True, ascending=False)
pecentage_success_fail.reset_index(inplace=True)

#format_dict = {'Pecentage Successful': "{:.2%}", 'Pecentage Failure':"{:.2%}"}

#pecentage_success_fail = pecentage_success_fail.style.format(format_dict).hide_index()

pecentage_success_fail
```

```python
pecentage_success_fail_plot = pecentage_success_fail

plt.figure(figsize=(20, 10), dpi=150)

x = pecentage_success_fail_plot['Company Name']
y = pecentage_success_fail_plot['Status Success']

plt.xlabel('Company Name')
plt.ylabel('Number of Successful Missions')
plt.title('Company with the morst successful missions')
plt.xticks(rotation=90)

plt.bar(x, y, fc='blue', ec= "black")

for i in range(len(x)):
    plt.text(i,y[i], y[i], ha= "center", va= "bottom")

plt.show()
```
